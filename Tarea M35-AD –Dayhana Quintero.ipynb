### Ticket promedio

```python
#Total por ticket
tickets = (df.groupby("Invoice ID", as_index=False).agg(ticket_total=("Total", "sum"),City=("City", "first"),Gender=("Gender", "first")))
tickets.head()
```

```python
#Ticket promedio por ciudad
ticket_promedio_ciudad = (tickets.groupby("City")["ticket_total"].mean())
ticket_promedio_ciudad.round(2)
```

```python
#Ticket promedio por product line y ciudad
ticket_pl_ciudad = (df.groupby(["Invoice ID", "City", "Product line"], as_index=False).agg(ticket_total=("Total", "sum")))
ticket_promedio_pl_ciudad = (ticket_pl_ciudad.groupby(["City", "Product line"])["ticket_total"].mean())
ticket_promedio_pl_ciudad.round(2)
```

```python
#Ticket promedio por género
ticket_promedio_genero = (tickets.groupby("Gender")["ticket_total"].mean())
ticket_promedio_ciudad.round(2)
```

```python
#Ticket promedio por genero y product line
ticket_genero_pl = (df.groupby(["Invoice ID", "Gender", "Product line"], as_index=False).agg(ticket_total=("Total", "sum")))
ticket_promedio_genero_pl = (ticket_genero_pl.groupby(["Gender", "Product line"])["ticket_total"].mean())
ticket_promedio_genero_pl.round(2)
```

- El ticket promedio se calcula tomando el monto total de cada factura (Invoice ID) y obteniendo el promedio de dichos montos según la categoría solicitada (ciudad, género, línea de producto). Se utiliza la columna Total porque representa el valor final pagado por el cliente, incluyendo impuestos.

### Top 5 ventas

```python
#Por ciudad
ventas_ciudad = df.groupby("City")["Total"].sum()
ventas_ciudad
```

```python
ranking_ciudad = ventas_ciudad.rank(ascending=False, method="dense")
ranking_ciudad
```

```python
top5_ciudad = (pd.DataFrame({"ventas": ventas_ciudad,"rank": ranking_ciudad})
    .query("rank <= 5").sort_values("ventas", ascending=False))
top5_ciudad
```

```python
total_ventas = df["Total"].sum()
top5_ciudad["aporte_%"] = (top5_ciudad["ventas"] / total_ventas) * 100
top5_ciudad.round(2)
```

```python
#Por customer
ventas_member = df.groupby("Customer type")["Total"].sum()
ranking_member = ventas_member.rank(ascending=False, method="dense")

top5_member = (pd.DataFrame({"ventas": ventas_member, "rank": ranking_member})
      .query("rank <= 5").sort_values("ventas", ascending=False))

top5_member["aporte_%"] = (top5_member["ventas"] / df["Total"].sum()) * 100
top5_member.round(2)
```

```python
#Por método de pago
ventas_payment = df.groupby("Payment")["Total"].sum()
ranking_payment = ventas_payment.rank(ascending=False, method="dense")

top5_payment = (pd.DataFrame({"ventas": ventas_payment, "rank": ranking_payment})
      .query("rank <= 5").sort_values("ventas", ascending=False))

top5_payment["aporte_%"] = (top5_payment["ventas"] / df["Total"].sum()) * 100
top5_payment.round(2)
```

### Matriz de correlación

```python
df["Hour"] = pd.to_datetime(df["Time"]).dt.hour
df[["Time", "Hour"]].head()
```

```python
corr_hora_total = df[["Hour", "Total"]].corr()
corr_hora_total
```

- La correlación entre la hora de la transacción y el total de venta es prácticamente nula (-0.0028), lo que indica que no existe una relación lineal significativa entre ambas variables. Esto sugiere que el monto de compra no está influenciado por la hora del día.

```python
corr_precio_rating = df[["Unit price", "Rating"]].corr()
corr_precio_rating
```

- La correlación entre el precio unitario y la calificación del cliente es prácticamente nula (-0.0088), lo que indica que el costo del producto no influye de manera significativa en la percepción del cliente medida a través del rating.
